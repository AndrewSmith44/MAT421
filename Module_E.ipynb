1. Continuity and Differentiation

Continuity and differentiation are fundamental concepts in calculus. A function is continuous if small changes in the input result in small changes in the output, and differentiable if it has a well-defined derivative.

Example Code: Checking Continuity

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.where(x < 0, x**2, x + 1)

x_values = np.linspace(-2, 2, 400)
y_values = f(x_values)

plt.plot(x_values, y_values, label=r'$f(x)$', color='blue')
plt.axvline(x=0, color='red', linestyle='--', label="x = 0")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Example of a Continuous Function')
plt.legend()
plt.show()
```

Example Code: Computing Derivatives Using SymPy

```python
import sympy as sp

x = sp.symbols('x')
f = x**3 + 2*x**2 - 5*x + 1

f_prime = sp.diff(f, x)
f_double_prime = sp.diff(f_prime, x)

print("First derivative:", f_prime)
print("Second derivative:", f_double_prime)
```

Example Code: Visualizing Taylor Approximation

```python
f_func = sp.lambdify(x, f, 'numpy')

x_values = np.linspace(-2, 2, 400)
y_values = f_func(x_values)

taylor_func = sp.lambdify(x, taylor_series.removeO(), 'numpy')
taylor_values = taylor_func(x_values)

plt.plot(x_values, y_values, label=r'$f(x)$', color='blue')
plt.plot(x_values, taylor_values, label='Taylor Approximation', linestyle='dashed', color='orange')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Taylor Series Approximation')
plt.legend()
plt.show()
```

3.1 Implementing Gradient Descent

Example Code: Gradient Descent on a Quadratic Function

```python
def f(x):
    return x**2 + 2*x + 1

def gradient(x):
    return 2*x + 2

learning_rate = 0.1
x_current = 5
iterations = 20

x_history = [x_current]
for _ in range(iterations):
    x_current = x_current - learning_rate * gradient(x_current)
    x_history.append(x_current)

x_values = np.linspace(-6, 4, 400)
y_values = f(x_values)

plt.plot(x_values, y_values, label=r'$f(x) = x^2 + 2x + 1$', color='blue')
plt.scatter(x_history, [f(x) for x in x_history], color='red', label='Gradient Descent Steps')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent Optimization')
plt.legend()
plt.show()
```
