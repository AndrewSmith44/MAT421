1. Interpolation Problem Statement

Interpolation allows us to estimate values within the range of a dataset by constructing new data points based on known values. This assignment will explore linear and cubic spline interpolation techniques using Python.

1.1 Dataset Overview

We will use the following dataset for all examples:

x	y

0	1

1	2

2	0

3	2

4	3

Python Code to visualize the dataset:

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.array([0, 1, 2, 3, 4])
y = np.array([1, 2, 0, 2, 3])

plt.scatter(x, y, color='red', label='Data Points')
plt.title("Interpolation Example Dataset")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

1.2 Why Interpolation?

Interpolation is used when we have discrete data points but need to estimate values between them.
Applications include image scaling, data smoothing, and physics simulations.
This assignment will implement two types of interpolation:
Linear interpolation for simplicity.
Cubic spline interpolation for smoother, more realistic curves.

2. Linear Interpolation

Linear interpolation estimates values between data points by connecting them with straight lines.

2.1 Python Code


```python
x = np.array([0, 1, 2, 3, 4])
y = np.array([1, 2, 0, 2, 3])

x_new = np.linspace(min(x), max(x), 50)
y_linear = np.interp(x_new, x, y)

x_combined = np.unique(np.concatenate((x, x_new)))
y_combined = np.interp(x_combined, x, y)

plt.scatter(x, y, color='red', label='Data Points', zorder=5)

plt.plot(x_combined, y_combined, color='blue', label='Linear Interpolation')

plt.title("Linear Interpolation (Aligned with Data Points)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

2.3 Advantages and Limitations
Advantages:

Simple to implement and computationally efficient.
Limitations:
Produces sharp angles at data points, which may not represent smooth real-world behavior.
3. Cubic Spline Interpolation

Cubic spline interpolation creates a smooth curve by fitting piecewise cubic polynomials between data points. It ensures continuity in both the first and second derivatives, providing a natural and smooth fit.

3.1 What is a Spline?

A spline is a piecewise polynomial function that can represent a curve.
In cubic spline interpolation, the function is continuous up to the second derivative, making the transitions between points smooth.

3.2 Python Implementation

We will use the scipy.interpolate.CubicSpline module to create and evaluate a cubic spline.

```python
from scipy.interpolate import CubicSpline


cs = CubicSpline(x, y)
y_cubic = cs(x_new)


plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x_new, y_cubic, color='green', label='Cubic Spline Interpolation')
plt.title("Cubic Spline Interpolation")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

3.3 Example Output

The cubic spline interpolation results in a smooth curve that passes through all data points, unlike the straight lines of linear interpolation.

4. Error Estimation and Discussion

While interpolation methods can approximate values effectively, errors may arise:

Extrapolation Issues: Both linear and cubic spline methods perform poorly outside the data range.

Data Quality: Inaccurate or sparse data points can lead to poor interpolation.

Computational Overhead: Cubic splines require more computation compared to linear interpolation.
